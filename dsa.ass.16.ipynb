# Given an array, for each element find the value of the nearest element to the right which is having a frequency greater than that of the current element. If there does not exist an answer for a position, then make the value â€˜-1â€™.

# **Examples:**

# Input: a[] = [1, 1, 2, 3, 4, 2, 1]
# Output : [-1, -1, 1, 2, 2, 1, -1]

# Explanation:
# Given array a[] = [1, 1, 2, 3, 4, 2, 1]
# Frequency of each element is: 3, 3, 2, 1, 1, 2, 3

# Lets calls Next Greater Frequency element as NGF
# 1. For element a[0] = 1 which has a frequency = 3,
#   As it has frequency of 3 and no other next element
#   has frequency more than 3 so  '-1'
# 2. For element a[1] = 1 it will be -1 same logic
#   like a[0]
# 3. For element a[2] = 2 which has frequency = 2,
#   NGF element is 1 at position = 6  with frequency
#   of 3 > 2
# 4. For element a[3] = 3 which has frequency = 1,
#   NGF element is 2 at position = 5 with frequency
#   of 2 > 1
# 5. For element a[4] = 4 which has frequency = 1,
#    NGF element is 2 at position = 5 with frequency
#   of 2 > 1
# 6. For element a[5] = 2 which has frequency = 2,
#    NGF element is 1 at position = 6 with frequency
#   of 3 > 2
# 7. For element a[6] = 1 there is no element to its
#   right, hence -1
# Input : a[] = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
# Output : [2, 2, 2, -1, -1, -1, -1, 3, -1, -1]

```python
def findNextGreaterFrequency(arr):
    frequency_map = {}
    result = []

    # Count the frequency of each element in the array
    for num in arr:
        if num in frequency_map:
            frequency_map[num] += 1
        else:
            frequency_map[num] = 1

    # Find the next greater frequency element for each element
    for i in range(len(arr)):
        found = False
        for j in range(i+1, len(arr)):
            if frequency_map[arr[j]] > frequency_map[arr[i]]:
                result.append(arr[j])
                found = True
                break
        if not found:
            result.append(-1)

    return result


# Test Example 1
arr = [1, 1, 2, 3, 4, 2, 1]
output = findNextGreaterFrequency(arr)
print(output)  # Output: [-1, -1, 1, 2, 2, 1, -1]

# Test Example 2
arr = [1, 1, 1, 2, 2, 2, 2, 11, 3, 3]
output = findNextGreaterFrequency(arr)
print(output)  # Output: [2, 2, 2, -1, -1, -1, -1, 3, -1, -1]
```

#  **Question 2**

# Given a stack of integers, sort it in ascending order using another temporary stack.

# **Examples:**

# Input : [34, 3, 31, 98, 92, 23]
# Output : [3, 23, 31, 34, 92, 98]

# Input : [3, 5, 1, 4, 2, 8]
# Output : [1, 2, 3, 4, 5, 8]

```python
def sortStack(stack):
    temp_stack = []

    while stack:
        # Remove the top element from the original stack
        temp = stack.pop()

        # Move elements from the temporary stack to the original stack
        while temp_stack and temp_stack[-1] > temp:
            stack.append(temp_stack.pop())

        # Push the current element to the temporary stack
        temp_stack.append(temp)

    # Move elements from the temporary stack back to the original stack
    while temp_stack:
        stack.append(temp_stack.pop())

    return stack


# Test Example 1
stack = [34, 3, 31, 98, 92, 23]
output = sortStack(stack)
print(output)  # Output: [3, 23, 31, 34, 92, 98]

# Test Example 2
stack = [3, 5, 1, 4, 2, 8]
output = sortStack(stack)
print(output)  # Output: [1, 2, 3, 4, 5, 8]
```

Given a stack withÂ **push()**,Â **pop()**, andÂ **empty()**Â operations, The task is to delete theÂ **middle**Â elementÂ ****of it without using any additional data structure.

Input Â :Â Stack[] = [1, 2, 3, 4, 5]

Output :Â Stack[] = [1, 2, 4, 5]

Input Â :Â Stack[] = [1, 2, 3, 4, 5, 6]

Output :Â Stack[] = [1, 2, 4, 5, 6]



```python
def deleteMiddle(stack):
    if len(stack) == 0:
        return

    n = len(stack)
    middle = n // 2

    # Helper function to delete the middle element recursively
    def deleteMiddleUtil(stack, k):
        # Base case: If the middle element is reached,
        # remove it from the stack
        if k == 0:
            stack.pop()
            return

        # Remove the top element and recur for the remaining stack
        temp = stack.pop()
        deleteMiddleUtil(stack, k - 1)

        # Push the top element back to the stack
        stack.append(temp)

    # Call the helper function to delete the middle element
    deleteMiddleUtil(stack, middle)


# Test Example 1
stack = [1, 2, 3, 4, 5]
deleteMiddle(stack)
print(stack)  # Output: [1, 2, 4, 5]

# Test Example 2
stack = [1, 2, 3, 4, 5, 6]
deleteMiddle(stack)
print(stack)  # Output: [1, 2, 4, 5, 6]
```

**Examples :**

Output :Â YesÂ

Input :Â Queue[] = { 5, 1, 2, 6, 3, 4 }Â 

Output :Â NoÂ 

Push 5 to stack.Â 

Pop 1, 2 from given Queue and push it to another Queue.Â 

Pop 6 from given Queue and push to stack.Â 

Pop 3, 4 from given Queue and push to second Queue.Â 

Now, from using any of above operation, we cannot push 5Â into the second Queue because it is below the 6 in the stack.



```python
def checkQueueSorting(queue):
    stack = []
    second_queue = []

    while not isEmpty(queue):
        front = dequeue(queue)

        if not stack:
            stack.append(front)
        else:
            if front < stack[-1]:
                stack.append(front)
            else:
                while stack and front > stack[-1]:
                    second_queue.append(stack.pop())
                stack.append(front)

    while stack:
        second_queue.append(stack.pop())

    while second_queue:
        enqueue(queue, second_queue.pop(0))

    return isSorted(queue)


# Function to check if a queue is empty
def isEmpty(queue):
    return len(queue) == 0


# Function to enqueue an element into the queue
def enqueue(queue, item):
    queue.append(item)


# Function to dequeue an element from the queue
def dequeue(queue):
    if isEmpty(queue):
        return None
    return queue.pop(0)


# Function to check if a queue is sorted in increasing order
def isSorted(queue):
    n = len(queue)
    if n <= 1:
        return True

    for i in range(n-1):
        if queue[i] > queue[i+1]:
            return False

    return True


# Test Example 1
queue = [5, 1, 2, 3, 4]
output = checkQueueSorting(queue)
print(output)  # Output: Yes

# Test Example 2
queue = [5, 1, 2, 6, 3, 4]
output = checkQueueSorting(queue)
print(output)  # Output: No
```

Given a number , write a program to reverse this number using stack.

**Examples:**

Input : 365
Output : 563

Input : 6899
Output : 9986

```python
def reverseNumber(num):
    stack = []
    reversed_num = ""

    # Convert the number to a string and push each digit onto the stack
    for digit in str(num):
        stack.append(digit)

    # Pop each digit from the stack to construct the reversed number
    while stack:
        reversed_num += stack.pop()

    return int(reversed_num)


# Test Example 1
num = 365
output = reverseNumber(num)
print(output)  # Output: 563

# Test Example 2
num = 6899
output = reverseNumber(num)
print(output)  # Output: 9986
```

Given an integer k and aÂ **[queue](https://www.geeksforgeeks.org/queue-data-structure/)**Â of integers, The task is to reverse the order of the firstÂ **k**Â elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

- **enqueue(x) :**Â Add an item x to rear of queue
- **dequeue() :**Â Remove an item from front of queue
- **size() :**Â Returns number of elements in queue.
- **front() :**Â Finds front item.


```python
from queue import Queue

def reverseFirstK(queue, k):
    stack = []
    
    # Remove the first k elements from the queue and push them onto the stack
    for _ in range(k):
        stack.append(queue.get())

    # Enqueue the elements from the stack back into the queue
    while stack:
        queue.put(stack.pop())

    # Rotate the remaining elements in the queue by k positions
    for _ in range(queue.qsize() - k):
        queue.put(queue.get())


# Test Example
queue = Queue()
queue.put(1)
queue.put(2)
queue.put(3)
queue.put(4)
queue.put(5)
queue.put(6)

k = 3

reverseFirstK(queue, k)

# Print the modified queue
while not queue.empty():
    print(queue.get(), end=" ")
```

Given a sequence of n strings, the task is to check if any two similar words come together and then destroy each other then print the number of words left in the sequence after this pairwise destruction.

**Examples:**

InputÂ : ab aa aa bcd ab

OutputÂ : 3

*As aa, aa destroys each other so,*

*ab bcd ab is the new sequence.*

Input : Â tom jerry jerry tom

Output :Â 0

*As first both jerry will destroy each other.*

*Then sequence will be tom, tom they will also destroy*

*each other. So, the final sequence doesnâ€™t contain any*

*word.*



```python
def destroySimilarWords(sequence):
    stack = []

    # Iterate through each word in the sequence
    for word in sequence.split():
        # If the stack is not empty and the top word is the same as the current word,
        # remove the top word from the stack (destroy both words)
        if stack and stack[-1] == word:
            stack.pop()
        # Otherwise, push the current word onto the stack
        else:
            stack.append(word)

    # Return the number of words left in the stack after pairwise destruction
    return len(stack)


# Test Example 1
sequence = "ab aa aa bcd ab"
output = destroySimilarWords(sequence)
print(output)  # Output: 3

# Test Example 2
sequence = "tom jerry jerry tom"
output = destroySimilarWords(sequence)
print(output)  # Output: 0
```

Given an array of integers, the task is to find the maximum absolute difference between the nearest left and the right smaller element of every element in the array.

**Note:**Â If there is no smaller element on right side or left side of any element then we take zero as the smaller element. For example for the leftmost element, the nearest smaller element on the left side is considered as 0. Similarly, for rightmost elements, the smaller element on the right side is considered as 0.

**Examples:**

Input : arr[] = {2, 1, 8}
Output : 1
Left smaller  LS[] {0, 0, 1}
Right smaller RS[] {1, 0, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 1

Input  : arr[] = {2, 4, 8, 7, 7, 9, 3}
Output : 4
Left smaller   LS[] = {0, 2, 4, 4, 4, 7, 2}
Right smaller  RS[] = {0, 3, 7, 3, 3, 3, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 7 - 3 = 4

Input : arr[] = {5, 1, 9, 2, 5, 1, 7}
Output : 1

```python
def maxAbsDifference(arr):
    n = len(arr)
    left_smaller = [0] * n
    right_smaller = [0] * n
    max_diff = float("-inf")

    # Find the nearest smaller element on the left side for each element
    for i in range(n):
        j = i - 1
        while j >= 0:
            if arr[j] < arr[i]:
                left_smaller[i] = arr[j]
                break
            j -= 1

    # Find the nearest smaller element on the right side for each element
    for i in range(n-1, -1, -1):
        j = i + 1
        while j < n:
            if arr[j] < arr[i]:
                right_smaller[i] = arr[j]
                break
            j += 1

    # Calculate the maximum absolute difference between left_smaller and right_smaller
    for i in range(n):
        diff = abs(left_smaller[i] - right_smaller[i])
        max_diff = max(max_diff, diff)

    return max_diff


# Test Example 1
arr = [2, 1, 8]
output = maxAbsDifference(arr)
print(output)  # Output: 1

# Test Example 2
arr = [2, 4, 8, 7, 7, 9, 3]
output = maxAbsDifference(arr)
print(output)  # Output: 4

# Test Example 3
arr = [5, 1, 9, 2, 5, 1, 7]
output = maxAbsDifference(arr)
print(output)  # Output: 1
```
